```python
print(err.keys())
```

```python
rep = classification_report(df["predict"], df["actual"], output_dict=True)

print("\nClassification Report:\n", rep)
```

```python
keys = list(rep.keys())

keys.remove('accuracy')
keys.remove('macro avg')
keys.remove('weighted avg')

print(keys)

print(len(keys))
```

```python
rec_dict = {i: rep[i]["recall"] for i in keys}

print(rec_dict)

sort_rec = {k: v for k, v in sorted(rec_dict.items(), key=lambda item: item[1])}

print(sort_rec)
```

```python
print(list(sort_rec.keys()))
```

```python
convert = json.load(open("/home/soumyajit/Project(-1)/DATASETS/data_let_it_wag/class_to_idx.json"))

# print(convert)

worst = list(sort_rec.keys())[:10]

name = [convert[i] for i in worst]

# print(worst)
# print(name)

for i in range(10):
    print(worst[i], name[i])
```

```python
name = sorted(name)


print(name)
```
